```python
import os
import getpass


def _set_env(var: str):
    if not os.environ.get(var):
        os.environ[var] = getpass.getpass(f"{var}: ")

_set_env("OPENAI_API_KEY")


import uuid
from flow_agent import FlowAgent, MockToolResult
agent = FlowAgent("plumbing_flow_wp.yaml")
```
